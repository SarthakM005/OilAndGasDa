```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\\Imp\\OilandGasDa\\DataFile\\Volve P-12_DatesCorrected.csv",index_col =0,parse_dates =True)
df.tail()
```

```python
#To develop a regression model that takes operational parameters as input and provides oil production rates as output
output_coloumn = 'BORE_OIL_VOL'
sns.heatmap(df.corr())# We find correlation between different operational paramters and remove either too high or too low correlation 
df.corr()
```

```python
#We only want to see correlation with the ouotput coloumn so plot it
df.corr()[output_coloumn].sort_values(ascending=False)[1:].plot(kind='bar', figsize=(12,4), grid=True) # The highly correlated params would be removed
#The bore gas volume is highly correlated to our output because they are produced simultaneously in the same pipe
corr_df = pd.DataFrame(df.corr()[output_coloumn].sort_values(ascending=False)[1:])
corr_df.columns = ['Corr']
corr_df
```

```python
#Now we select our input features
input_columns =corr_df[(abs(corr_df['Corr'])>0.2) & (corr_df['Corr']<0.80)].index
input_columns

X = df[input_columns] ; y = df[output_coloumn]
X.shape , y.shape
```

```python
#Now forecasting
df_model_preparation = df[df.index.year <=2014]
df_model_testing = df[df.index.year>=2015]
df_model_preparation.head()
df_original = df_model_preparation.copy()
df = df_original.loc[:,['BORE_OIL_VOL']]
print(df)
df.plot(figsize=(12,5))
```

```python
# What you do is , u shift the data down and then use a loop to keep on shifting the data down
df.shift(1)
```

```python
window = 60

def TS_to_Xy(df, n_lags=3, window=window):
    df = df.copy()  # Avoid modifying the original DataFrame
    df = df.rolling('30D').mean().dropna()  # Apply rolling mean and drop NaNs

    for i in range(1, n_lags + 1):
        df[f'y_{i}'] = df['BORE_OIL_VOL'].shift(i + window)  # Shift only the target column

    return df.dropna()  # Drop remaining NaNs

df_Xy = TS_to_Xy(df, 1)  
print(df_Xy.head())  # Ensure data is valid

```

```python
df_Xy = df_Xy.fillna(method='bfill')
df_Xy['y_1'].plot(figsize=(12,5))   
```

```python
from sklearn.model_selection import train_test_split
X = df_Xy[[col for col in df_Xy.columns if col.startswith('y')]]
y = df_Xy['BORE_OIL_VOL']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42,shuffle=False)
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
yp_train = rf.predict(X_train)
```

```python
plt.figure(figsize=(12,4))

plt.plot(X_train.index, y_train)
plt.plot(X_train.index, yp_train)

plt.title(f'RMSE = {np.sqrt(mean_squared_error(y_train,yp_train))}', size=20)
```

```python
ti = 800;  tf = 1500 

plt.figure(figsize=(15,5))

plt.scatter(df_Xy.index[ti:tf] , df_Xy['BORE_OIL_VOL'][ti:tf], marker='o', label='Actual Oil Rates', s=200)
plt.plot(df_Xy.index[ti:tf] , rf.predict(df_Xy[['y_1']][ti-window:tf-window]),color='k', label='Predicted Rates')


plt.plot(df_Xy.index[ti:tf] , df_Xy['BORE_OIL_VOL'][ti:tf], marker='x', label='Actual Oil Rates')

plt.grid()
plt.legend()
```
